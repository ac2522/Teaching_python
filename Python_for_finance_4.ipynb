```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
    
chipo = pd.read_csv(url, sep = '\t')

%matplotlib inline
```

```python
# we have a dataset and we want to draw a scatter graph
# between price of order and items ordered per order_id 

chipo
# in the data each row is an item
# each item has an order_iid and a price
# for each order_id we want to combine all the quantities
# and all prices
```

```python
print(chipo.item_price[0])
print(type(chipo.item_price[0]))
# first issue is that you cant add strings
print(chipo.item_price[0]+ chipo.item_price[1])

# so we need to turn this into a number
```

```python
chipo["item_price2"] = [float(value[1:-1]) for value in chipo.item_price] 
# strip the dollar sign and trailing space
# for each value in chip.item_price (value is arbitrary name)
# chipo.item_price = [float(spkl[1:-1]) for spkl in chipo.item_price] works
# for each value take the second place in the string to the penultimate
# chipo.item_price[0][1:-1]
# and then convert into number using function float()
# imprtant to remember that int and float are different object types
# if i replaced float with int() it would round down

# the fact that it is in [ ] saves it as a list 
# it then creates new column item_price2 and converts list into panda series
```

```python
chipo.item_price = [float(value[1:-1]) for value in chipo.item_price] 
chipo.drop(["item_price2"],axis=1,inplace=True)
```

```python
orders = chipo.groupby('order_id').sum()
# groups dataframe  by order_id then sums the orders
# so that you get total number of items per order_id and total price per id
```

```python
plt.scatter(x = orders.item_price, y = orders.quantity, s = 30, c = 'blue')
#plt.scatter takes two series with same number of items
#i think it can also take lists
# s=30 changes size of dots 

```

```python
plt.scatter(x = orders.item_price, y = orders.quantity, s = 10, c = 'red')
plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price')
plt.ylim(0)
plt.show
# the way plt works is that is prints graph at end of code
# or when you write plt.show
```

```python
# so if i do multiple graphs they will appear together

plt.scatter(x = orders.item_price, y = orders.quantity, s = 30, c = 'blue')
plt.scatter(x = orders.item_price, y = orders.quantity, s = 8, c = 'red')
plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price')
plt.ylim(0)
plt.show
```
