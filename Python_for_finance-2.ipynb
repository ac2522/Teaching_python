## ALgorithm complexity

```python
"""The time complexity of an algorithm is the total amount of time required by an algorithm to complete its execution.
In simple words, every piece of code we write, takes time to execute. 
The time taken by any piece of code to run is known as the time complexity of that code.
The lesser the time complexity, the faster the execution."""
```

```python
def constant(n): # this function adds 3 to the input, and then divides by two returning the new number
    n = (n + 3) / 2
    return n

print(constant(5)) # processes here are adding 5 and 3, dividing by 2, and assigning n = ...
# that is three
print(constant(600)) # here it is the same

# different processes take different operations take different amounts of time. i.e ** and +
# i.e. 183-2+16/4 has 3 operations  23**67 is 1 operation yet you can imagine it will take more time

# so what we do is generalize the terms
# if there are 1000 operations but that doesnt change we say that it is O(1) which means constant time
# the function above is constant, because it doesnt matter the size of n,
# it will take approximately the same amount of time
```

```python
# a function of linear time will generalize to O(n)

def linear(n):
    for i in range(n):  # remember the function range() which will iterate through 0 up to the number n
        x = i + 3
        print(x)
    return

# here the function linear will take input size n
```

```python
O(n**3)
```

```python
O(1000+1000000)
```

```python
linear(5)
# inputting 5 will make it print out 5 numbers

# to get the complexity look through 
# it starts with i=0:
# 1) it assigns i to be 0
# 2) it adds 3
# 3) it assigns to x
# 4) it prints x

# thats 4, this will repeat n times
# therefore the algorithmic complexity is 4n

# however we generalize this to n
# so complexity is O(n)
```

```python
#if you dont know what log(n) is: imagine I give you a problem:
# I am thnking of a number bigger than 0 and less than n  which we will say is 100
# I will tell you if it is bigger than or smaller than my number after each guess
# How do you choose?
# if you try 1 then 2, then 3 chances are you will guess it after 50 attempts (or n/2)
# this is O(n) - linear

# however hopefully you noticed that if you first say 50 I will tell you if it is > or <
# so you can remove half the numbers as options
# and then between 50 and 100 you can guess 75, and then youve halfed the playing field again.
```

```python
def is_x_in_list(list_, x):
    found = False ; b = 0 ; n = len(list_) - 1 
    while found == False:
        halfway = (n + b) // 2
        print(b,'-',n)
        if list_[halfway] == x:
            return halfway
        elif halfway == n:
             return "not here"
        elif list_[halfway] > x:
            n = halfway -1
        elif list_[halfway] < x:
            b = halfway+1
```

```python
is_x_in_list(list(range(10000)),11)
print('______________________________________')
is_x_in_list(list(range(10000000)),11)
# does increase as n increases but not linearly
```

```python
def difficult2(n):
    sum = 0
    for i in range(1,n+1):
        for j in range(i^2):
            if j % i == 0:
                for k in range(j):
                    sum+=1
    return sum
```

```python
def fibonacci(n): # difficult
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
def get_into_safe(n): # safe is n numbers long
    iterations = 1
    for x in range(n):
        iterations *= 10
    for y in range(iterations):
        if break_safe(y) == True:
            return y
    raise Exception('We didnt get in trt again later')
    
def break_safe(y):
    if y== 3876:      # 3876 is current key so change this if you want to change key
        return True
    else: 
        return False
```
