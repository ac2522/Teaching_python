```python
def distance_between_postcodes(postcode1, postcode2):  # distance between post_codes
    if type(postcode1) == str: 
        postcode1 = postcode1.replace(' ', '').lower()
    else:
        return -1
    
    if type(postcode2) == str:
        postcode2 = postcode2.replace(' ', '').lower()
    else:
        return -2
    
    if postcode1 and postcodes_info.index.contains(postcode1) \
            and postcode2 and postcodes_info.index.contains(postcode2):
        pc1_info = postcodes_info.loc[postcode1.replace(' ', '').lower()]
        pc2_info = postcodes_info.loc[postcode2.replace(' ', '').lower()]

        slat = radians(float(pc1_info['latitude']))  # input("Starting latitude: ")
        slon = radians(float(pc1_info['longitude']))  # input("Ending longitude: ")
        elat = radians(float(pc2_info['latitude']))  # input("Starting latitude: ")
        elon = radians(float(pc2_info['longitude']))  # input("Ending longitude: ")

        dist = 6371.01 * acos(sin(slat)*sin(elat) + cos(slat)*cos(elat)*cos(slon - elon))

        return dist
    elif not postcodes_info.index.contains(postcode1):
        return -3  # error cases
    elif not postcodes_info.index.contains(postcode2):
        return -4  # error cases


def grouper(n, iterable, fillvalue=None):
    "grouper(3, 'ABCDEFG', 'x') --> ABC DEF Gxx"
    args = [iter(iterable)] * n
    return zip_longest(fillvalue=fillvalue, *args)


results = []
unique_postcodes = []
for a in range(len(df_c)):
    pc1 = df_c["Postal Code"][a]
    pc2 = df_c["Postcode"][a]
    unique_postcodes.append(pc1)
    unique_postcodes.append(pc2)
    unique_postcodes = list(set(unique_postcodes))

for pcs in tqdm(list(grouper(90, unique_postcodes))):   #range((unique_postcodes.shape[0] // 100)+1)): the bar at bottom
    st = {"postcodes": pcs}
    r = requests.post(url='http://api.postcodes.io/postcodes', data=st)  # api gets data from postcodes
    results.extend(r.json()['result'])
	

    postcodes_info = pd.DataFrame(results)
    postcodes_info = postcodes_info.apply(lambda x: pd.Series(x['result']), axis=1)
    postcodes_info = postcodes_info.dropna(subset=['postcode']).drop_duplicates(subset=['postcode'])
    postcodes_info = postcodes_info.set_index(postcodes_info['postcode'].str.replace(' ', '').str.lower())
```

```python
def word_share(sen1, sen2):
    score = 0
    if (type(sen1) == str) and (type(sen2) == str):
        sen1 = sen1.lower().split()
        sen2 = sen2.lower().split()
        for s in sen1:
            if s in sen2:
                score += 1
    return score

def word_share2(sen1, sen2):
    if (type(sen1) == str) and (type(sen2) == str):
        if sen1 in sen2:
            return 1
    return 0
    
df_cd2['talent'] = df_cd2.apply(lambda x: word_share(x['Job talent pools'],x['Talent Pools']), axis=1)
df_cd2['emploi'] = df_cd2.apply(lambda x: word_share2(x['Employment type'],x['Preferred employment']), axis=1)

df_cd2.drop(['Job talent pools', 'Talent Pools','Employment type','Preferred employment'],axis=1, inplace=True)
```
